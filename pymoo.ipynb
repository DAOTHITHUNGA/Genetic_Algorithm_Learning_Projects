```
from pymoo.algorithms.nsga2 import NSGA2
from pymoo.algorithms.moead import MOEAD
from pymoo.factory import get_problem, get_visualization, get_reference_directions, get_termination, get_sampling, get_crossover, get_mutation, get_performance_indicator
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
import numpy as np
```

```
def run_NSGA2(problem, pop_size=100, n_evals=20000, verbose=0):
  algorithm = NSGA2(pop_size=pop_size)
  termination = get_termination("n_eval", n_evals)
  return minimize(problem, algorithm, termination, seed=18520750, verbose=verbose)

def run_MOEAD(problem, pop_size = 91, n_evals=20000,n_neighbors=15, n_dims = 3, n_partitions=12, verbose=0):
  algorithm = MOEAD(
    get_reference_directions("energy", n_dims, n_partitions, seed=18520750),
    pop_size = pop_size,
    n_neighbors=n_neighbors,
    decomposition="pbi",
    prob_neighbor_mating=0.7,
    seed=18520750
  )
  termination = get_termination("n_eval", n_evals)
  return minimize(problem, algorithm, seed=18520750, termination=termination, verbose=verbose)

def plot_(problem, result, type='zdt'):
  # The pareto front of a scaled zdt1 problem
  if type == 'zdt':
    pf = problem.pareto_front()
  elif type == 'dasmop':
    pf = problem
  # The result found by an algorithm
  A = result.F

  # plot the result
  return Scatter(legend=True).add(pf, label="Pareto-front").add(A, label="Result")
```

```
problem_set = ['zdt1', 'zdt2', 'zdt3', 'zdt4', 'zdt6','dascmop1','dascmop2','dascmop3','dascmop4','dascmop5','dascmop6']
```

#ZDT1

```
problem = get_problem(problem_set[0])
res_moe = run_MOEAD(problem, pop_size=200, n_evals=20000, n_dims=3, n_neighbors=15, n_partitions=200)
res_nsga2 = run_NSGA2(problem, pop_size=200, n_evals=20000)
pf = problem.pareto_front()

gd = get_performance_indicator("gd", pf)
gd_plus = get_performance_indicator("gd+", pf)
igd = get_performance_indicator("igd", pf)
igd_plus = get_performance_indicator("igd+", pf)

print("MOEA/D - GD", gd.calc(res_moe.F))
print("MOEA/D - GD+", gd_plus.calc(res_moe.F))
print("MOEA/D - IGD", igd.calc(res_moe.F))
print("MOEA/D - IGD+", igd_plus.calc(res_moe.F))

print("\nNSGAII - GD", gd.calc(res_nsga2.F))
print("NSGAII - GD+", gd_plus.calc(res_nsga2.F))
print("NSGAII - IGD", igd.calc(res_nsga2.F))
print("NSGAII - IGD+", igd_plus.calc(res_nsga2.F))

graph_moe = plot_(problem, res_moe)
graph_nsga2 = plot_(problem, res_nsga2)

graph_moe.show()
graph_nsga2.show()
```
